```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
training_features = pd.read_csv(r"C:\Users\Akshay Phatak\Documents\Atharva\ML\IITGuwati_Datascience\dataset and all\training_set_features.csv")
training_features.head()
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
```

```python
categorical_features = ['age_group', 'education', 'race', 'sex', 'income_poverty', 'marital_status', 'rent_or_own', 'employment_status', 'hhs_geo_region', 'census_msa','employment_industry', 'employment_occupation']
numerical_features = [col for col in training_features.columns if col not in categorical_features]
```

```python
numerical_features
```

```python
scaler = StandardScaler()
training_features[numerical_features] = scaler.fit_transform(training_features[numerical_features])
```

```python
training_features.head()
```

```python
one_hot_encoder = OneHotEncoder(sparse=False, drop='first')
encoded_features = one_hot_encoder.fit_transform(training_features[categorical_features])
encoded_df = pd.DataFrame(encoded_features, columns=one_hot_encoder.get_feature_names_out(categorical_features))
training_features= pd.concat([training_features.drop(columns=categorical_features), encoded_df], axis=1)
```

```python
for i in training_features.columns.to_list():
    print(i, training_features[i].isna().sum())
    print(i, training_features[i].isnull().sum())
    print("")
```

```python
training_features= training_features.fillna(training_features.mean())
```

```python
training_labels = pd.read_csv(r"C:\Users\Akshay Phatak\Documents\Atharva\ML\IITGuwati_Datascience\dataset and all\training_set_labels.csv")
```

```python
training_labels_xyz = training_labels['xyz_vaccine']
training_labels__seasonal = training_labels['seasonal_vaccine']
```

```python
from sklearn.model_selection import train_test_split
X_train_xyz, X_test_xyz, y_train_xyz, y_test_xyz = train_test_split(training_features,training_labels_xyz, test_size=0.21, random_state=21)
```

```python
y_train_xyz
```

```python
RFC = RandomForestClassifier()
RFC.fit(X_train_xyz,y_train_xyz)
```

```python
y_pred_xyz_RFC = RFC.predict_proba(X_test_xyz)
```

```python
y_pred_xyz_RFC
```

```python

print(RFC.classes_)
```

```python
y_pred_xyz__RFC1 = [y_pred_xyz_RFC[i][1] for i in range(len(y_pred_xyz_RFC))]
```

```python
from sklearn.metrics import *

roc_auc_RFC = roc_auc_score(y_test_xyz, y_pred_xyz__RFC1, average='macro')


print(f'ROC AUC Score: {roc_auc_RFC}')
```

```python
from sklearn import svm

svm_c = svm.SVC(probability=True)
```

```python
svm_c.fit(X_train_xyz,y_train_xyz)
y_pred_xyz_SVM = svm_c.predict_proba(X_test_xyz)
```

```python
y_pred_xyz_SVM
```

```python
y_pred_xyz__SVM1 = [y_pred_xyz_SVM[i][1] for i in range(len(y_pred_xyz_SVM))]
roc_auc_SVM = roc_auc_score(y_test_xyz, y_pred_xyz__SVM1, average='macro')
print(f'ROC AUC Score: {roc_auc_SVM}')
```

```python
from sklearn.linear_model import LogisticRegression
LR= LogisticRegression()
LR.fit(X_train_xyz,y_train_xyz)
y_pred_xyz_LR = LR.predict_proba(X_test_xyz)


y_pred_xyz__LR1 = [y_pred_xyz_LR[i][1] for i in range(len(y_pred_xyz_LR))]
roc_auc_LR = roc_auc_score(y_test_xyz, y_pred_xyz__LR1, average='macro')
print(f'ROC AUC Score: {roc_auc_LR}')
```

```python
from sklearn.naive_bayes import GaussianNB
NB = GaussianNB()
NB.fit(X_train_xyz,y_train_xyz)
y_pred_xyz_NB = NB.predict_proba(X_test_xyz)


y_pred_xyz__NB1 = [y_pred_xyz_NB[i][1] for i in range(len(y_pred_xyz_NB))]
roc_auc_NB = roc_auc_score(y_test_xyz, y_pred_xyz__NB1, average='macro')
print(f'ROC AUC Score: {roc_auc_NB}')
```

```python
ROC_AUC_curves = [roc_auc_LR,roc_auc_NB,roc_auc_RFC,roc_auc_SVM]
ROC_AUC_Models = ['Logistic Regression', 'NaiveBayse', 'Random Forest Classifier', 'Support Vector Machine']

sns.barplot(ROC_AUC_curves)
plt.xlabel(ROC_AUC_Models)
plt.yticks()
plt.ylim((0.7,0.9))
```
